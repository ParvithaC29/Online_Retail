```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
```

```python
data['Year'] = data['InvoiceDate'].dt.year
data['Month'] = data['InvoiceDate'].dt.month
data['Day'] = data['InvoiceDate'].dt.day
data['Hour'] = data['InvoiceDate'].dt.hour
data['Minute'] = data['InvoiceDate'].dt.minute
```

```python
data['TotalSales'] = data['Quantity'] * data['UnitPrice']
profit_margin = 0.20
data['Profit'] = data['TotalSales'] * profit_margin
data['TotalSales'] = data['TotalSales'].round(2)
data['Profit'] = data['Profit'].round(2)
pd.set_option('display.float_format', lambda x: f'{x:,.2f}')
```

```python
import sqlite3
```

```python
conn = sqlite3.connect(":memory:")
```

```python
data.to_sql("retail", conn, index=False, if_exists="replace")
```

```python
query = """
SELECT SUM(TotalSales) AS TotalRevenue,
       SUM(Profit) AS TotalProfit,
       COUNT(DISTINCT InvoiceNo) AS TotalOrders
FROM retail;
"""
overview = pd.read_sql(query, conn)
print("\n Sales Overview:\n", overview)
overview[['TotalRevenue', 'TotalProfit']].plot(kind='bar', color=['skyblue', 'orange'])
plt.title("Total Revenue vs Total Profit")
plt.ylabel("Amount")
plt.show()
```

```python
query = """
SELECT Description, SUM(TotalSales) AS TotalSales
FROM retail
GROUP BY Description
ORDER BY TotalSales DESC
LIMIT 10;
"""
top_products = pd.read_sql(query, conn)
print("\n Top 10 Selling Products:\n", top_products)
sns.barplot(data=top_products, x="TotalSales", y="Description")
plt.title("Top 10 Selling Products")
plt.show()
```

```python
query = """
SELECT Country, SUM(Profit) AS TotalProfit
FROM retail
GROUP BY Country
ORDER BY TotalProfit DESC
LIMIT 10;
"""
top_countries = pd.read_sql(query, conn)
print("\n Top 10 Countries by Profit:\n", top_countries)
sns.barplot(data=top_countries, x="TotalProfit", y="Country")
plt.title("Top 10 Countries by Profit")
plt.show()
```

```python
query = """
SELECT Year, Month, SUM(TotalSales) AS TotalSales
FROM retail
GROUP BY Year, Month
ORDER BY Year, Month;
"""
monthly_sales = pd.read_sql(query, conn)
print("\n Monthly Sales Trend:\n", monthly_sales)
sns.lineplot(data=monthly_sales, x="Month", y="TotalSales", hue="Year", marker="o")
plt.title("Monthly Sales Trend")
plt.show()
```

```python
query = """
SELECT Hour, SUM(TotalSales) AS TotalSales
FROM retail
GROUP BY Hour
ORDER BY Hour;
"""
hourly_sales = pd.read_sql(query, conn)
print("\n Hourly Sales Pattern:\n", hourly_sales)
sns.barplot(data=hourly_sales, x="Hour", y="TotalSales")
plt.title("Hourly Sales Pattern")
plt.show()
```

```python
query = """
SELECT CustomerID, SUM(Profit) AS TotalProfit
FROM retail
WHERE CustomerID != 0
GROUP BY CustomerID
ORDER BY TotalProfit DESC
LIMIT 10;
"""
top_customers = pd.read_sql(query, conn)
print("\n Top 10 Customers by Profit:\n", top_customers)
sns.barplot(data=top_customers, x="TotalProfit", y="CustomerID")
plt.title("Top 10 Customers by Profit")
plt.show()
```

```python
query = """
SELECT
    CASE
        WHEN Month IN (1,2,3) THEN 'Q1'
        WHEN Month IN (4,5,6) THEN 'Q2'
        WHEN Month IN (7,8,9) THEN 'Q3'
        WHEN Month IN (10,11,12) THEN 'Q4'
    END AS Quarter,
    SUM(TotalSales) AS TotalSales
FROM retail
GROUP BY Quarter
ORDER BY Quarter;
"""
quarter_sales = pd.read_sql(query, conn)
print("\n Seasonal Sales by Quarter:\n", quarter_sales)
sns.barplot(data=quarter_sales, x="Quarter", y="TotalSales")
plt.title("Seasonal Sales by Quarter")
plt.show()
```

```python
query_aov = """
SELECT Country, AVG(TotalSales) AS AvgOrderValue, COUNT(DISTINCT InvoiceNo) AS NumOrders
FROM retail
GROUP BY Country
ORDER BY AvgOrderValue DESC
LIMIT 10;
"""
aov_country = pd.read_sql(query_aov, conn)
aov_country['AvgOrderValue'] = aov_country['AvgOrderValue'].round(2)
print("\n Top 10 Countries by Average Order Value:\n", aov_country)
plt.figure()
sns.barplot(data=aov_country, x='AvgOrderValue', y='Country')
plt.title("Top 10 Countries by Average Order Value")
plt.xlabel("Average Order Value")
plt.tight_layout()
plt.show()
```

```python
basket = data.groupby('InvoiceNo')['Quantity'].sum()
avg_basket_size = basket.mean()
median_basket_size = basket.median()
print(f"\n Basket Size:\n Average items per order = {avg_basket_size:.2f}\n Median items per order = {median_basket_size:.2f}")
plt.figure()
sns.histplot(basket, bins=50, kde=False)
plt.title("Distribution of Basket Sizes (Items per Order)")
plt.xlabel("Number of Items in Invoice")
plt.ylabel("Count of Invoices")
plt.tight_layout()
plt.show()
```
